```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/AI4Good Project/PD_concat_dataset.csv")
df.rename(columns={'Vowel_type (NaN=not a vowel, 0="a")':'Vowel_type', 'Status (0=control, 1=PD)':'Status'}, inplace=True)
df.drop(df[df['Vowel_type'] != 0.0].index, inplace=True) # remove recordings that aren't vowels
print(df.columns)
```

```python
features = ['motor_UPDRS', 'total_UPDRS', 'GenderSex', 
            'MDVP:Jitter(Abs)',
            'Jitter:PPQ5', 'MDVP:Shimmer', 'MDVP:Shimmer(dB)', 
            'Shimmer:APQ3', 'Shimmer:APQ5', 'Shimmer:APQ11', 
            'RPDE', 'DFA', 'PPE']

# generate heatmap
plt.figure(figsize=(18, 8))
sns.heatmap(df[features].corr(), vmin = -1, vmax = 1, annot = True, cmap = 'coolwarm')
plt.title('Correlation Map Of Parkinson\'s Features', fontdict={'fontsize':12}, pad=12)
```

```python
# generate boxplots

for feature in features[3:]:
  control = df.loc[df.Status == 0, feature]
  park = df.loc[df.Status == 1, feature]
  data = [control, park]

  fig, ax = plt.subplots()
  ax.set_title(feature)
  ax.boxplot(data, autorange=True)
  plt.show()
```
